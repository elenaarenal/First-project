# Read me

BICIMAD is the municipal bicycle rental service of the city of Madrid. It has 264 stations (2024 update, it has about 600) spreaded over diferent districts of the city.  The aim of this project is allowing the potential users to find the nearest BiciMAD station to a set of non-municipal accessible facilities of the city of Madrid (considered places of interest)using an app called **BicimadLocator**

### Status

Since this is the first Ironhack Data Analytics Project , the code is in its initial stages, therefore it will be modified and updated. For now on, the only functionality is the abovementioned but it will be bidirectional ( allowing searching for bm station as well) and the code it will be improved using FuzzyWuzzy (to help locate the places/stations even if the provided name is not exactly as defined in code)

### Technology stack

For the development of this project, the following libraries/imports have been used:

    pandas
    numpy
    json
    requests
    geopandas
    argparse
    
As data sources, we have used:

1)API  from cityhall containing all the accesible non-municipal facilities, its location, and other data that we found not relevant for the purpose of this Project.
url="https://datos.madrid.es/egob/catalogo/202180-0-instalaciones-accesibles-no-muni.json"

         
2)Bicimad csv file (not updated) containing all the bicimad stations, their location,  their operational status and other data that we found not relevant for the purpose of this Project.
It is included in the folder called Data ('bicimad_stations.csv). Be aware that columns shall be separated by tab (sep='\t')

**Aspects that have been considered**:

  The csv of bicimad contains fixed data, therefore it is not updated with the current info: the availability of bikes has not been considered, only the operatibility of the station itself
  
  Codification UTF-8 is requiered as the first draft of the table shows a high number of error characters. Eg:  Leter ñ in Nuestra señora de la Almudena
  
   Both files have been searched for null values. 
   The location have been split into two columns latitude and longitud to include these parameters in the geolocalization functions.
   
 **Data matching**:  
 A new csv file has been creared containing  the following data extracted from both data sources:
All the accesible non-municipal facilities, its location (longitude/latitude) matched with all the bicimad stations, it’s adress and location(longitude/latitude) and the distance between the two points. The distance have been calculated using geopandas and it has been optimized leaving only the minimum distance for each par of values. **Be aware that this steep it tooks about 10 minuts when run** 




 
        




### How to use it (argparse)

The file to be executed it's called: argbicimad.py

Execute trough the terminal( python argbicimad.py) including the following input:

Input'-t', '--titulo’, 

help='Instalación accesible no municipal, input todos/as para listado completo') 

*OPTION 1*(destination point)
For the user who wants to get the closest bicimad Station to a specific place:
The user shall input the name of the place of interest to be reached.
E.g: “Museo del Prado”


*OPTION 2*
For the user who wants to get the information for all the places of interest, the input shall be “todos” (the string todos is not case sensitive)
E.g: “TODAS”


### Data obtained

A CSV file it's exported. The outoput table called Bicimadfinal2 shows the following information:

-Non municipal accesible facility

-The closest bicimad station and it exact address

-The distance betwen the two points (straight-line distance)
